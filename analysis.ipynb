```python
import pandas as pd

from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure, output_notebook, show
from bokeh.palettes import Spectral6
from bokeh.transform import factor_cmap
from bokeh.palettes import brewer
from IPython.display import display

output_notebook()
```

```python
df = pd.read_csv(r"../data/results_with_manual_classification.csv")
```

```python
# Get the number of colors we'll need for the plot.
colors = brewer["Spectral"][len(df.delay.unique())]

# Create a map between factor and color.
colormap = {delay:colors[idelay] for idelay, delay in enumerate(df.delay.unique())}

# Create a list of colors for each value that we will be looking at.
colors = [colormap[x] for x in df.delay]
```

```python
df['color']= df['delay'].apply(lambda x:colormap[x])
tools = ["box_zoom", "pan", "hover", "reset"]

p = figure(plot_width=700, plot_height=400, tools=tools)

source = ColumnDataSource(data=df.loc[df['delay']=='130ns', :])
p.circle('PIV_intensity_debris', 'debris_contrast', 
         source=source, color='black', legend='delay')
show(p)
```

```python
p = figure(plot_width=700, plot_height=400, tools=tools)

source = ColumnDataSource(data=df.loc[(df['delay']=='130ns')&(df['PIV_intensity_debris']), :].reset_index().sort_values(by='PIV_intensity_debris'))
p.circle('index', 'PIV_intensity_debris', 
         source=source, color='black')
show(p)
```
